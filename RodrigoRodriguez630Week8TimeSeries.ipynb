```python
# Reshape the dataset from wide to long format used for time series data
data_long = pd.melt(df, id_vars=['YEAR'], var_name='MONTH', value_name='SALES')
```

```python
# Convert YEAR and MONTH to a datetime format
data_long['DATE'] = pd.to_datetime(data_long['YEAR'].astype(str) + data_long['MONTH'], format='%Y%b')
```

```python
# Sort by DATE
data_long = data_long.sort_values(by='DATE')

# Drop columns we no longer need
data_long = data_long.drop(['YEAR', 'MONTH'], axis=1).set_index('DATE')
```

**#1 Plot the data with proper labeling and make some observations on the graph.**

```python
plt.figure(figsize=(14, 7))
plt.plot(data_long.index, data_long['SALES'], label='Monthly Retail Sales')
plt.title('Monthly Retail Sales in the US (1992-2021)')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Show summary of the data to spot any missing values or anomalies
data_long.describe(), data_long.isna().sum()
```

```python
data_long.tail(12)
```

```python
# Drop missing values
data_longV2 = data_long.dropna(axis=0)
```

```python
# Looking for the anomoly in 2020
data_longV2.tail(24)
```

**Observations**

So far we can see a clear increase in sales over time with a drop in sales occuring from 2008 and 2009.

Seeing as the drop in sales coincides with the financial housing crisis of 2008, we could assume that this drop was due to the recession.

Additonally, there is an anomoly in April 2020 where sales took a sharp dip only to immediately bounce back the following month.

**#2. Split this data into a training and test set. Use the last year of data (July 2020 â€“ June 2021) of data as your test set and the rest as your training set.**

```python
# Split the data into training and test sets using the datasets without NaN values
train = data_longV2.loc[:'2020-06']
test = data_longV2.loc['2020-07':'2021-06']
```

```python
# Verifing the splits by checking the start and end dates of each dataset
train_start, train_end = train.index.min(), train.index.max()
test_start, test_end = test.index.min(), test.index.max()

train_start, train_end, test_start, test_end
```

**#3. Use the training set to build a predictive model for the monthly retail sales.**

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Build and fit the Holt-Winters model
model = ExponentialSmoothing(train['SALES'], trend='add', seasonal='add', seasonal_periods=12).fit()
```

**#4. Use the model to predict the monthly retail sales on the last year of data.**

```python
# Predict on the training set to visualize how well the model fits the data
train_predictions = model.fittedvalues

# Predict on the test set
test_predictions = model.forecast(steps=len(test))
```

```python
# Plot the training data, test data, and forecasts
plt.figure(figsize=(14, 7))
plt.plot(train.index, train['SALES'], label='Training Data')
plt.plot(test.index, test['SALES'], label='Actual Test Data')
plt.plot(test.index, test_predictions, label='Forecasted Test Data', linestyle='--')
plt.title('Train and Test Data with Forecasts')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

**#5. Report the RMSE of the model predictions on the test set.**

```python
from sklearn.metrics import mean_squared_error

# Calculate RMSE
rmse = mean_squared_error(test['SALES'], test_predictions, squared=False)

rmse
```

The RMSE is high in my opinion. However, I think the big rmse might just be due to the macro economic events of the pandemic. The market saw a sharp drop initially then some companies saw large demand fueled by the pandemic.

The case about Chase Bank in our textbook states that predictive models can not predict macro-economic events and I think that's what we are seeing here. However, choosing a better model or tuning hyperparameters would of course improve performance.
