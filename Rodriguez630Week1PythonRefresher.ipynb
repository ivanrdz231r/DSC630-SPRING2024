### *Plots*

*Reading and opening the data*

```python
import pandas as pd

# reading the data
df_hs = pd.read_csv('C:/Users/ivan2/gitLocal/DSC550-WINTER2023/campus-data-download-4yr-longitudinal-2018-2020-class.csv.csv')
df_hs.head(5)
```

**1)** *Creating a histogram plot to answer the first problem question.*

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid") # Set the style of seaborn
```

```python
# Create the histogram for Graduation Rates
grad_rates = df_hs.CAMP_ALLR_GRAD
plt.hist(grad_rates, bins=50, label='Graduation Rates')
plt.xlabel('All Schools Graduation Rates')
plt.ylabel('Frequency')
plt.title('Histogram of Graduation Rates')

plt.legend()
plt.show()
```

**2)** *Creating a box plot to answer the second problem question.*

```python
# Create a box plot for Graduation Rates by Region Type
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(x='REGION_TYPE', y='CAMP_ALLR_GRAD', data=df_hs, palette={1: "salmon", 2: "slateblue"})

# manually setting x-axis tick labels
boxplot.set_xticklabels(['Rural', 'Urban'])

# titles and labels
boxplot.set_title('Grad Rates by Region Type')
boxplot.set_xlabel('Region Type')
boxplot.set_ylabel('CAMP_ALLR_GRAD')

# Show the plot
plt.show()
```

**3)** Creating a box plot to answer the third problem question.

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
regression_plot = sns.regplot(x='CAMP_ALLR_GRAD', y='CAMP_ALLD', data=df_hs, scatter_kws={'alpha':0.5})

# Set the title and labels
regression_plot.set_title('Relationship between Class Size and Graduation Rate')
regression_plot.set_xlabel('Graduation Rate (CAMP_ALLR_GRAD)')
regression_plot.set_ylabel('Class Size (CAMP_ALLD)')

# Show the plot
plt.show()
```

**Summary**

The first histogram plot shows us that graduation rates are left-skewed with most graduation rates being above 80%. Additionally we can see some outliers showing a graduation rate of 0%. This will have to be cleaned during data prep.

The second box plot shows us that on average Rural region type graduation rates are higher than Urban region type. Again we can see some outliers with graduation rates of 0% that need to be cleaned.

The third plot shows the relationship between class size and graduation rate. The trend line's small positive slope suggests that there is a weak positive relationship between class size and graduation rate. As class size increases, there is a very slight tendency for the graduation rate to increase as well. However, given the slope is small, this relationship is not particularly strong.
